```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('weatherHistory.csv')
df.head()
```

```python
df = df.drop_duplicates()
print("Kích thước dữ liệu sau khi xóa dòng trùng:", df.shape)
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True, errors='coerce')
df['Formatted Date'].head()
```

```python
df['Year'] = df['Formatted Date'].dt.year
df['Month'] = df['Formatted Date'].dt.month
df['Day'] = df['Formatted Date'].dt.day
df['Hour'] = df['Formatted Date'].dt.hour

df[['Formatted Date', 'Year', 'Month', 'Day', 'Hour']].head()
```

```python
print(df.isnull().sum())
```

```python
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Đã xử lý thiếu cho cột dạng chuỗi")
print(df[cat_cols].isnull().sum())
```

```python
num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

print("Đã xử lý thiếu cho cột số")
print(df[num_cols].isnull().sum())
```

```python
label_encoders = {}
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

df.head()
```

```python
df.to_csv('weather_cleaned.csv', index=False)
print("Đã lưu file weather_cleaned.csv")
```
