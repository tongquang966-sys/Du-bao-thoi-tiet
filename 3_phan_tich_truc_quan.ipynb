```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('weather_cleaned.csv')
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Temperature (C)'], bins=30, kde=True)
plt.title('Phân bố nhiệt độ')
plt.xlabel('Temperature (C)')
plt.ylabel('Tần suất')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Humidity'], bins=30, kde=True)
plt.title('Phân bố độ ẩm')
plt.xlabel('Humidity')
plt.ylabel('Tần suất')
plt.show()
```

```python
monthly_temp = df.groupby('Month')['Temperature (C)'].mean()

plt.figure(figsize=(10,5))
monthly_temp.plot(marker='o')
plt.title('Nhiệt độ trung bình theo tháng')
plt.xlabel('Tháng')
plt.ylabel('Nhiệt độ trung bình (C)')
plt.grid(True)
plt.show()
```

```python
monthly_humidity = df.groupby('Month')['Humidity'].mean()

plt.figure(figsize=(10,5))
monthly_humidity.plot(marker='o')
plt.title('Độ ẩm trung bình theo tháng')
plt.xlabel('Tháng')
plt.ylabel('Độ ẩm trung bình')
plt.grid(True)
plt.show()
```

```python
yearly_temp = df.groupby('Year')['Temperature (C)'].mean()

plt.figure(figsize=(10,5))
yearly_temp.plot(marker='o')
plt.title('Nhiệt độ trung bình theo năm')
plt.xlabel('Năm')
plt.ylabel('Nhiệt độ trung bình (C)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Chỉ lấy các cột số
df_num = df.select_dtypes(include=['number'])

plt.figure(figsize=(12,8))
sns.heatmap(df_num.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Ma trận tương quan giữa các biến số')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Temperature (C)'])
plt.title('Biểu đồ boxplot của nhiệt độ')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Humidity'], y=df['Temperature (C)'])
plt.title('Mối quan hệ giữa độ ẩm và nhiệt độ')
plt.xlabel('Humidity')
plt.ylabel('Temperature (C)')
plt.show()
```
